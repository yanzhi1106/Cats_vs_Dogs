```python
import tensorflow as tf
import tensorflow.keras as keras
import tensorflow.keras.layers as layers
import cv2 as cv
import numpy as np
```

```python
# 获取训练好的模型
model = keras.models.load_model('/content/drive/MyDrive/Cat_vs_Dog/model4.h5')
```

```python
# 导入验证集数据
evalData = np.load('/content/drive/MyDrive/Cat_vs_Dog/evalDataNormalized.npy')
evalLabel = np.load('/content/drive/MyDrive/Cat_vs_Dog/evalLabelOneHot.npy')
```

```python
# 使用 model.evaluate 函数执行模型评估。
result = model.evaluate(evalData, evalLabel)
```

```python
# 使用单张图片测试
def preparePredict(path):
    image = cv.imread(path)
    image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    image = cv.resize(image, (224, 224))
    image = (image - 128.0) / 128.0
#模型支持一次性推理多张图像，因此我们需要在 image 变量之前新增一个维度
#使用 np.expand_dims 函数对图像进行扩增,该函数需要传入 2 个参数，分别为待扩增维度的数据 image 和扩增的位置 axis.
    image = np.expand_dims(image, axis = 0)
    return image
```

```python
#调用 preparePredict 函数分别处理 test1.jpg 和 test2.jpg 图像。
test1 = preparePredict('/content/drive/MyDrive/Cat_vs_Dog/cat/cat.1103.jpg')
test2 = preparePredict('/content/drive/MyDrive/Cat_vs_Dog/dog/dog.51.jpg')
```

```python
# 把两个图片合并成一个值
x = np.concatenate((test1, test2), axis = 0)
```

```python
# 做模型推理
yPredict = model.predict(x)
```
